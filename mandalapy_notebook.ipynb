## ** MANDALAS IN PYTHON **

## This is a python implementation of fronkonstin's R code for drawing mandalas 

Change the function _plotutils.py from scipy spatial functions. In the line 225 of the code (highlighted on the link), change: linestyle = 'dashed' to linestyle = 'solid', and save the function.

```python
#Import the necesary packages
#The source code of the function 'voronoi_plot_2d' has been modified
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from math import pi
```

###  Initial paramaeters (They can be modified as you like)
If you want a different figure from the default, just change this parameters and rerun the next cells

```python
it = 4 #Number of iterations (how many times the equidistant points will be generated)
points = 7 #Number of points to draw
radius = 2 #Distance between points, increasing it will expand the final figure
```

### Set the angles between points, specify the initial center, obtain voronoi regions and plot


```python
angles = np.linspace(0,2*pi*(1-1/points), points) + pi/2 #Angles between points 
x,y = 0,0
df = pd.DataFrame([[x,y]], columns = ['x','y']) #Initial center

#Itearate over centers
for k in range(it):
    #Initialize empty numpy arrays for appending coordinates
    t1 = np.array([])
    t2 = np.array([])
    for i in range(df.shape[0]):
        t1 = np.append(t1,df['x'][i]+radius**(k)*np.cos(angles))
        t2 = np.append(t2,df['y'][i]+radius**(k)*np.sin(angles))
    df = pd.DataFrame(np.column_stack((t1,t2)), columns = ['x','y'])
    
## Obtain voronoi regions
mandala = Voronoi(df)
plt.clf()
#Set the size of the figure
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] ,fig_size[1] = 20,20 #Change the previous 2 numbers if you want a bigger or smaller figure 
#Draw the mandala and eliminate all unnecesary elements from the figure 
voronoi_plot_2d(mandala, show_points = False, show_vertices = False), plt.axis('off')
plt.show()
```

### Want to save the figure?

```python
#Change the name if you want(default:'my_mandala.png')
voronoi_plot_2d(mandala, show_points = False, show_vertices = False), plt.axis('off'), plt.savefig('fmandala2.png', dpi = 100)
```
